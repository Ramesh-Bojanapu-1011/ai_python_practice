```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv("homeprices.csv")
df
```

Label Encoding

```python
le = LabelEncoder()
```

```python
df.town = le.fit_transform(df.town)
df
```

Onehot Encoding

```python
# Select only the "town" column from the DataFrame.
# Using double brackets keeps it as a DataFrame instead of a Series.
X = df[["town"]]

# Create a ColumnTransformer that will apply OneHotEncoder
# only to the first (and only) column of X.
ct = ColumnTransformer(
    [("town", OneHotEncoder(), [0])]  # (name, transformer, column index)
)

# Fit the transformer on the "town" data and convert it into
# a one-hot encoded matrix.
X = ct.fit_transform(X)

# View the transformed output (usually a sparse matrix).
X
```

```python
# Remove the first column from the one-hot encoded matrix
# (usually to avoid the dummy variable trap).
X = X[:, 1:]
X
```
