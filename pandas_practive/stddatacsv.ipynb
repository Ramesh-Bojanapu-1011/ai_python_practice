```python
import pandas as pd
```

```python
stddatacsv = pd.read_csv("stddatacsv.csv")
stddatacsv_df = pd.DataFrame(stddatacsv)
stddatacsv_df
```

```python
# Drop columns where all values are NaN (i.e., completely empty columns)
stddatacsv_df.dropna(axis=1, how="all", inplace=True)
stddatacsv_df
```

```python
# Drop rows where all values are NaN (i.e., completely empty rows)
stddatacsv_df.dropna(how="all", inplace=True)
stddatacsv_df
```

```python
# Remove duplicate rows from the DataFrame, keeping only the first occurrence of each duplicate
stddatacsv_df.drop_duplicates(inplace=True)
stddatacsv_df
```

```python
# Count the number of missing (NaN) values in each column of the DataFrame
stddatacsv_df.isnull().sum()
```

```python
stddatacsv_df.address.isnull().sum()  # Count missing values in 'address' column
```

```python
stddatacsv_df.address = stddatacsv_df.address.fillna(
    stddatacsv_df.address.mode()[0]
)  # Fill missing 'address' values with the mode (most frequent value)
stddatacsv_df.address.isnull().sum()
```

```python
stddatacsv_df.m1 = stddatacsv_df.m1.fillna(
    round(stddatacsv_df.m1.mean(), 1)
)  # Fill missing 'm1' values with the mean rounded to 1 decimal place
stddatacsv_df.m3 = stddatacsv_df.m3.fillna(
    round(stddatacsv_df.m3.mean(), 1)
)  # Fill missing 'm3' values with the mean rounded to 1 decimal place
stddatacsv_df.m4 = stddatacsv_df.m4.fillna(
    round(stddatacsv_df.m4.mean(), 1)
)  # Fill missing 'm4' values with the mean rounded to 1 decimal place
stddatacsv_df[["m1", "m3", "m4"]].isnull().sum()
```

```python
stddatacsv_df.isna().sum()
```

```python
stddatacsv_df
```

```python
stddatacsv_df.dropna(
    thresh=5, inplace=True
)  # If a row has 5 or fewer non-null values, it will be removed.
stddatacsv_df
```

```python
# Checks the columns 'm3' and 'm4' only.
# If either column has NaN, that entire row will be removed.
# All other columns are ignored for this operation.

stddatacsv_df.dropna(subset=["m3", "m4"], inplace=True)
stddatacsv_df
```

```python
stddatacsv_df.drop(
    ["regno", "phone"], axis=1, inplace=True
)  # Drop 'regno' and 'phone' columns
stddatacsv_df
```
