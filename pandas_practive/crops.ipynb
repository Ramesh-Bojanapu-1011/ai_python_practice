```python
import pandas as pd
```

```python
crops = pd.read_csv("crops.csv")
crops_df = pd.DataFrame(crops)
crops_df
```

```python
df2=crops_df.sample(n=18) # Randomly selects 18 records from the dataframe
df2
```

```python
df3 = crops_df.loc[51:75,['crop','year','area']]
print(df3) # Gives records from 51 - 75 for 3 columns
```

```python
crops_df.loc[51:69] # Gives records from 51 - 69 for all columns
```

```python
crops_df.loc[84:,['crop','year','area']] # Gives records from 84 to end for 3 columns
```

```python
crops_df.iloc[10:20,[0,3,4]] # Gives rows 10–19 and only specific columns: 0, 3, and 4
```

```python
crops_df.iloc[10:20,1:4] # Gives rows 10 to 19 and columns 1 to 3 using integer positions
```

```python
crops_df.iloc[:,[0,2,4]] # Gives all rows and only specific columns: 0, 2, and 4
```

```python
crops_df.iloc[:20] # Gives first 20 records for all columns
```

```python
crops_df.iloc[:,-1] # Gives all rows for the last column
```
