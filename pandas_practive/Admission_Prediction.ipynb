```python
df.isna().sum()  # checking for missing values
```

```python
df["GRE Score"] = df["GRE Score"].fillna(
    round(df["GRE Score"].mean())
)  # filling missing values with mean
df["GRE Score"].isnull().sum()
```

```python
df["TOEFL Score"] = df["TOEFL Score"].fillna(
    round(df["TOEFL Score"].mean())
)  # filling missing values with mean
df["TOEFL Score"].isnull().sum()
```

```python
df["University Rating"] = df["University Rating"].fillna(
    df["University Rating"].mode()[0]
)  # filling missing values with mode
df["University Rating"].isnull().sum()
```

```python
df.isna().sum()
```

```python
df.to_csv("admission_Prediction_cleaned.csv")
```
