```python
import pandas as pd
```

```python
movie_metadata = pd.read_csv(
    "movie_metadata.csv", sep=","
)  # Reading the CSV file with semicolon separator
movie_metadata_df = pd.DataFrame(movie_metadata)  # Converting to DataFrame
movie_metadata_df
```

```python
dt = movie_metadata_df[
    movie_metadata_df["duration"] > 170
]  # Filtering movies with duration greater than 170 minutes
dt
```

```python
data = movie_metadata_df[
    ["title_year", "imdb_score"]
]  # Selecting specific columns: title_year and imdb_score
data
```

```python
movie_metadata_df.columns
```

```python
# Filter movies that:
# 1) have duration > 160 minutes,
# 2) have fewer than 100 critic reviews,
# 3) are produced in the USA
data2 = movie_metadata_df[
    (movie_metadata_df.duration > 160)
    & (movie_metadata_df.num_critic_for_reviews < 100)
    & (movie_metadata_df.country == "USA")
]
data2
```

```python
# Filter movies with duration between 160 and 200 minutes,
# sort them by duration in ascending order,
# and display only the director_name, movie_title, and duration columns
data3 = (
    movie_metadata_df[
        (movie_metadata_df.duration >= 160) & (movie_metadata_df.duration < 200)
    ]
    .sort_values(by="duration")
    .reset_index(drop=True)
)
data3[["director_name", "movie_title", "duration"]]
```

```python
# Sort the movie metadata by duration in descending order (longest movies first)
data4 = movie_metadata_df.sort_values(by="duration", ascending=False)
data4[["movie_title", "duration"]]
```

```python
data5 = movie_metadata_df[
    movie_metadata_df.genres.str.contains("Comedy")
]  # Filtering movies in the Comedy genre
data5[["movie_title", "genres"]]
```

```python
# Filter movies that are either:
# 1) belong to both "Action" and "Thriller" genres, or
# 2) belong to the "Drama" genre
data6 = movie_metadata_df[
    movie_metadata_df.genres.str.contains(r"(?=.*Action)(?=.*Thriller)")
    | movie_metadata_df.genres.str.contains("Drama")
]
# Display only the movie title and genres columns
data6[["movie_title", "genres"]]
```

```python
movie_metadata_df['country'].head()
```

```python
# Get the number of occurrences of each unique country,
movie_metadata_df['country'].value_counts() 
```

```python
# Replace missing values in the 'country' column with the most frequent country (mode)
movie_metadata_df['country'].fillna(movie_metadata_df['country'].mode()[0], inplace=True)

# Display the first 5 values of the 'country' column to verify the changes
movie_metadata_df['country'].head()
```

```python
movie_metadata_df['duration'].head()
```

```python
movie_metadata_df['duration'].isnull().sum()  # Count of missing values in 'duration' column
```

```python
# Fill missing values in the 'duration' column with the mean duration (rounded to 1 decimal place)
movie_metadata_df['duration'].fillna(round(movie_metadata_df.duration.mean(), 1), inplace=True)

# Display the first 5 values of the 'duration' column
movie_metadata_df['duration'].isnull().sum() 
```
