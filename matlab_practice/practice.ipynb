```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [65, 97, 38, 81, 47, 73, 59]
```

```python
plt.bar(x, y)
for i in range(len(x)):
    plt.text(x[i], y[i], int(y[i]))

plt.title("sample bar graph")
plt.xlabel("students")
plt.ylabel("marks")
plt.legend(["marks"])
plt.show()
```

```python
plt.plot(x, y)
for i in range(len(x)):
    plt.text(x[i], y[i], int(y[i]), va="bottom")

plt.title("sample line graph")
plt.xlabel("students")
plt.ylabel("marks")
plt.legend(["marks"])
plt.show()
```

```python
plt.scatter(x, y)
for i in range(len(x)):
    plt.text(x[i], y[i], int(y[i]))
    
plt.title("sample scatter graph")
plt.xlabel("students")
plt.ylabel("marks")
plt.legend(["marks"])
plt.show()
```

```python
plt.barh(x, y)
for i in range(len(x)):
    plt.text(y[i], x[i], int(y[i]))
plt.title("sample horizontal bar graph")
plt.xlabel("marks")
plt.ylabel("students")
plt.legend(["marks"])
plt.show()
```
