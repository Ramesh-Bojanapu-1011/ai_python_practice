```python
import matplotlib.pyplot as plt
```

```python
mobilemarket = ["samsung", "apple", "oneplus", "moto", "nokia"]
sizes = [200, 127, 86, 155, 34]
```

```python
colors = ["r", "b", "y", "g", "c"]
explode = [0, 0, 0.5, 0, 0.25]
```

```python
plt.pie(sizes, labels=mobilemarket, autopct="%.1f")
plt.title("Mobile market")
plt.show()
```

```python
plt.pie(
    sizes,
    labels=mobilemarket,
    colors=colors,
    startangle=0,
    shadow=True,
    explode=explode,
    autopct="%.2f",
)
plt.title("Mobile market")
plt.show()
```

```python
bank = ["sbi", "icici", "hdfc", "axis", "union"]
loan = [300, 255, 179, 192, 78]
```

```python
colors = ["r", "b", "k", "g", "y"]
explode = (0, 0, 0, 0, 0)
```

```python
plt.pie(
    loan,
    labels=bank,
    colors=colors,
    startangle=50,
    shadow=True,
    explode=explode,
    autopct="%.1f",
)
plt.title("Bank loans")
plt.show()
```
