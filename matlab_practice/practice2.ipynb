```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "Year": [2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
        "Sales": [2000, 2300, 4800, 3500, 6000, 5200, 4900, 6300],
    }
)
df
```

If we create graph by the using of datafarame we dont need plt.legend()


```python
df.plot("Year", "Sales", kind="line", label="sales per year", color="red")
plt.title("Simple Line Plot")
plt.xlabel("Year section")
plt.ylabel("Sales as per year")
plt.show()
```

```python
df.plot("Year", "Sales", kind="bar", label="sales per year", color="red")
plt.title("Simple Line Plot")
plt.xlabel("Year section")
plt.ylabel("Sales as per year")
plt.show()
```

```python
df.plot("Year", "Sales", kind="scatter", label="sales per year", color="red")
plt.title("Simple Line Plot")
plt.xlabel("Year section")
plt.ylabel("Sales as per year")
plt.show()
```

If we create graph by the without using of datafarame we need plt.legend()


```python
plt.plot(df.Year, df.Sales, label="sales per year", color="red")
plt.title("Simple Line Plot")
plt.xlabel("Year section")
plt.ylabel("Sales as per year")
plt.legend()
plt.show()
```

```python
plt.bar(df.Year, df.Sales, label="sales per year", color="red")
plt.title("Simple Line Plot")
plt.xlabel("Year section")
plt.ylabel("Sales as per year")
plt.legend()
plt.show()
```

Marker Line Chart


Color codes for stying of graph
| Letter | Color Name |
| ------ | ---------- |
| `b` | Blue |
| `g` | Green |
| `r` | Red |
| `c` | Cyan |
| `m` | Magenta |
| `y` | Yellow |
| `k` | Black |
| `w` | White |


```python
df.plot(
    "Year",
    "Sales",
    kind="line",
    title="Simple Line Plot",
    label="sales per year",
    style="g--o",  # 'color:line_style:marker'
    # style examples: 'g--o', 'r:D', 'b-.^', 'k:x', 'm-.*'
)

plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

multi line plot for two lines


```python
product = pd.DataFrame(
    {
        "Year": [2017, 2018, 2019, 2020, 2021, 2022],
        "ProdASales": [2000, 3000, 4000, 3500, 6000, 5300],
        "ProdBSales": [3000, 4000, 3500, 3500, 5500, 6100],
    }
)
product
```

```python
ax1 = product.plot(
    "Year", "ProdASales", kind="line", label="Product A Sales", style="r-*"
)
product.plot(
    "Year",
    "ProdBSales",
    ax=ax1,
    kind="line",
    label="Product B Sales",
    title="MultiLine Plot",
    style="b--o",
)  # ax : axes object
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

multi line plot for Three lines


```python
product2 = pd.DataFrame(
    {
        "Year": [2017, 2018, 2019, 2020, 2021, 2022],
        "ProdASales": [2000, 3000, 4000, 3500, 6000, 5300],
        "ProdBSales": [3000, 4000, 3500, 3500, 5500, 6100],
        "ProdCSales": [3900, 2000, 3100, 5000, 4500, 4200],
    }
)
product2
```

```python
product2.plot(
    "Year",
    ["ProdASales", "ProdBSales", "ProdCSales"],
    kind="line",
    title="MultiLine Plot for 3 Products",
    style=["r-*", "b--o", "g-.^"],
)
plt.xlabel("Year")
plt.ylabel("Sales")
plt.show()
```

```python
ICICIBank_Loans = pd.DataFrame(
    {
        "Year": [2014, 2015, 2016, 2017, 2018, 2019, 2020],
        "Loans in C": [7.5, 25, 15, 38, 6, 12, 20],  # add two values
        "Returns in c": [1.5, 9, 11, 27.7, 0.3, 3.2, 4.0],  # add two values
        "Profit in c": [0.5, 5.4, 3.1, 7, 9, 2.5, 6.0],  # add two values
    }
)
ICICIBank_Loans
```

```python
ICICIBank_Loans.plot(
    "Year",
    ["Loans in C", "Returns in c", "Profit in c"],
    kind="line",
    title="ICICI Bank Loans, Returns and Profit Analysis",
    style=["c-*", "k--o", "b-.^"],
)
plt.xlabel("Year")
plt.ylabel("Amount in C")
plt.show()
```
