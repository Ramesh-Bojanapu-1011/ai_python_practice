```python
import matplotlib.pyplot as plt
```

```python
x = ["Fathima", "Abhi", "Esha", "Balu", "Divya", "Chandu"]
y = [57, 93, 29, 78, 84, 61]
```

```python
c1 = ["red", "red", "red", "green", "green", "green"]
```

```python
# sort the data based on Y-val
y, x = zip(*sorted(zip(y, x)))  # sorted order
```

```python
plt.bar(x,y,label="sorted data")
for i in range(len(x)):
    plt.text(x[i], y[i]+1, str(y[i]), ha='center')
plt.legend()
plt.title("sample bar graph")
plt.xlabel("marks")
plt.ylabel("students")
plt.show()
```

```python
# plt.bar(x,y,label="sorted data")
plt.bar(x, y, color=c1, label="sorted data")  # sorted order
for i in range(len(x)):
    plt.text(x[i], y[i]+1, str(y[i]), ha='center')
plt.legend()
plt.title("sample bar graph")
plt.xlabel("marks")
plt.ylabel("students")
plt.show()
```
