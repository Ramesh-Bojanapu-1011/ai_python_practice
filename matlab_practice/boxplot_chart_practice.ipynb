```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.normal(10, 50, 45)
data
```

```python
plt.boxplot(data)
plt.show()
```

```python
feature_1 = np.random.normal(10, 100, 200)
feature_2 = np.random.normal(10, 100, 200)
feature_3 = np.random.normal(10, 100, 200)
feature_4 = np.random.normal(10, 100, 200)
```

```python
data = [feature_1, feature_2, feature_3, feature_4]
data
```

```python
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data, tick_labels=["feature_1", "feature_2", "feature_3", "feature_4"])
plt.show()
```

```python
plt.boxplot(
    data,
    patch_artist=True,
    tick_labels=["feature_1", "feature_2", "feature_3", "feature_4"],
)
plt.show()
```

```python
plt.boxplot(
    data,
    notch=True,
    patch_artist=True,
    tick_labels=["feature_1", "feature_2", "feature_3", "feature_4"],
)
plt.show()
```
