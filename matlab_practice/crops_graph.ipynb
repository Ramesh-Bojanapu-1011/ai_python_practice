```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crops = pd.read_csv("crops.csv")
crops_df = pd.DataFrame(crops)
data=crops_df.head(20)
data
```

```python
crops_df.columns
```

```python
data.plot("year", "yields", kind="line", marker="o")
plt.title("Crops Production Over Years")
plt.xlabel("Year")
plt.ylabel("Production (in million tons)")
plt.show()
plt.clf()
```

```python
data.plot("year", "yields", kind="bar", color="orange")
plt.title("Crops Production Over Years")
plt.ylabel("Year")
plt.xlabel("Production (in million tons)")
plt.show()

```
